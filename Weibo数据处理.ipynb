```python
model = gensim.models.word2vec.Word2Vec.load('word2vec.model')

def predict_proba(oword, iword):
    iword_vec = model[iword]
    oword = model.wv.vocab[oword]
    oword_l = model.syn1[oword.point].T
    dot = np.dot(iword_vec, oword_l)
    lprob = -sum(np.logaddexp(0, -dot) + oword.code*dot) 
    return lprob

from collections import Counter
def keywords(s):
    s = [w for w in s if w in model]
    ws = {w:sum([predict_proba(u, w) for u in s]) for w in s}
    return Counter(ws).most_common()

s = u'新冠疫情'
pd.Series(keywords(jieba.cut(s)))
```

```python
text_copy_4 = text_undu
topic1_list = []
topic2_list = []
topic3_list = []
topic4_list = []
topic5_list = []
temp_list = []
for index, row in text_copy_4.iterrows():
    temp = pd.Series(keywords(jieba.cut(row['text'])))
    temp_list.append(temp)
    length = len(temp)
    if length == 0:
        topic1 = '无'
        topic2 = '无'
        topic3 = '无'
        topic4 = '无'
        topic5 = '无'
    elif length == 1:
        topic1 = temp[0][0]
        topic2 = '无'
        topic3 = '无'
        topic4 = '无'
        topic5 = '无'
    elif length == 2:
        topic1 = temp[0][0]
        topic2 = temp[1][0]
        topic3 = '无'
        topic4 = '无'
        topic5 = '无'
    elif length == 3:
        topic1 = temp[0][0]
        topic2 = temp[1][0]
        topic3 = temp[2][0]
        topic4 = '无'
        topic5 = '无'
    elif length == 4:
        topic1 = temp[0][0]
        topic2 = temp[1][0]
        topic3 = temp[2][0]
        topic4 = temp[3][0]
        topic5 = '无'
    elif length >= 5:
        topic1 = temp[0][0]
        topic2 = temp[1][0]
        topic3 = temp[2][0]
        topic4 = temp[3][0]
        topic5 = temp[4][0]
    topic1_list.append(topic1)
    topic2_list.append(topic2)
    topic3_list.append(topic3)
    topic4_list.append(topic4)
    topic5_list.append(topic5)
```

```python
#主题词聚类
topics_list = []
for index, row in text_undu.iterrows():
    temp_topics = row['topic1'] + ' ' + row['topic2'] + ' ' + row['topic3'] + ' ' + row['topic4'] + ' ' + row['topic5']
    if '无' in temp_topics:
        temp_topics = list(set(temp_topics)).remove('无')
    topics_list.append(temp_topics)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vec=CountVectorizer()

km_matrix= count_vec.fit_transform(res)

print(km_matrix.shape)
```

```python
print(km_matrix.toarray())
```

```python
from sklearn.cluster import KMeans

num_clusters = 4 #聚为四类，可根据需要修改

km = KMeans(n_clusters=num_clusters)

km.fit(km_matrix)

clusters = km.labels_.tolist()
```
